#lenguaje phyton crear la clase "animal" con los atributos "nombre", "raza" constructor animal y las funciones ladrar() correr() crear los objetos pitbull chihuahua nivel principiente no es interactivo , proporciona tu los datos

```python
# Definición de la clase Animal
class Animal:
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    def ladrar(self):
        print(f"{self.nombre} dice: ¡Guau guau! 🐾")

    def correr(self):
        print(f"{self.nombre} está corriendo velozmente... 🏃‍♂️💨")

    # Nueva función: Acción de morder
    def morder_vecina(self):
        print(f"⚠️ ¡CUIDADO! {self.nombre} ha mordido a la vecina. 👵🚑")

# --- Creación de los objetos ---

# 1. El Pitbull
pitbull = Animal("Rocky", "Pitbull 🐶")

# 2. El Chihuahua
chihuahua = Animal("Fifi", "Chihuahua 🐕")

# 3. El Perro Callejero
callejero = Animal("Firulais", "Callejero (Sultán de la calle) 🦴")

# --- Ejecución de acciones ---

print("--- REPORTE CANINO ---")

# Datos del Pitbull
print(f"Nombre: {pitbull.nombre} | Raza: {pitbull.raza}")
pitbull.ladrar()
print("-" * 30)

# Datos del Chihuahua
print(f"Nombre: {chihuahua.nombre} | Raza: {chihuahua.raza}")
chihuahua.correr()
print("-" * 30)

# Datos del Callejero y el incidente con la vecina
print(f"Nombre: {callejero.nombre} | Raza: {callejero.raza}")
callejero.ladrar()
callejero.morder_vecina() # Aquí activamos la función solicitada
```
